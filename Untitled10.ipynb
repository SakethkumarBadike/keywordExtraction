```python
import json
import os
from transformers import AutoTokenizer
from datasets import Dataset


MODEL_NAME = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)


with open('ner_ready_data.json', 'r', encoding='utf-8') as f:
    raw_data = json.load(f)

# PRE-FILTER: Remove any entries that have no text or are empty
raw_data = [item for item in raw_data if item.get('text') and len(str(item['text']).strip()) > 0]

# --- STEP 1: CREATE BIO LABEL MAPPINGS ---
unique_labels = set()
for entry in raw_data:
    for entity in entry.get('entities', []):
        unique_labels.add(entity[2])

label2id = {"O": 0}
id_counter = 1

for label in sorted(list(unique_labels)):
    label2id[f"B-{label}"] = id_counter
    label2id[f"I-{label}"] = id_counter + 1
    id_counter += 2

id2label = {v: k for k, v in label2id.items()}

# --- STEP 2: TOKENIZE AND ALIGN WITH BIO LOGIC ---
def tokenize_and_align_labels(entry, max_length=512):
    text = str(entry.get('text', ""))
    entities = entry.get('entities', [])

    tokenized_input = tokenizer(
        text,
        truncation=True,
        max_length=max_length,
        return_offsets_mapping=True,
        padding="max_length"
    )

    # Initialize labels with 'O' and handle special tokens
    labels = [label2id["O"]] * len(tokenized_input["input_ids"])
    offsets = tokenized_input["offset_mapping"]

    # Sort entities by start character
    entities.sort(key=lambda x: x[0])

    for start_char, end_char, label_name in entities:
        if f"B-{label_name}" not in label2id:
            continue

        b_label = label2id[f"B-{label_name}"]
        i_label = label2id[f"I-{label_name}"]

        first_token_found = False
        for idx, (start, end) in enumerate(offsets):
            # Mask special tokens ([CLS], [SEP], [PAD])
            if start == 0 and end == 0:
                labels[idx] = -100
                continue

            # Check if token falls within the entity range
            if start >= start_char and end <= end_char:
                if not first_token_found:
                    labels[idx] = b_label
                    first_token_found = True
                else:
                    labels[idx] = i_label

    tokenized_input["labels"] = labels
    return tokenized_input

# --- STEP 3: PROCESS AND SAVE ---
processed_data = []
for i, item in enumerate(raw_data):
    try:
        processed_data.append(tokenize_and_align_labels(item))
    except Exception as e:
        print(f"Skipping index {i} due to error: {e}")

full_dataset = Dataset.from_list(processed_data)
dataset_dict = full_dataset.train_test_split(test_size=0.2, seed=42)

dataset_dict.save_to_disk("processed_resume_dataset_bio")

with open("ner_config_bio.json", "w") as f:
    json.dump({"label2id": label2id, "id2label": id2label, "model": MODEL_NAME}, f, indent=4)

print(f"✅ BIO Dataset ready! Processed {len(processed_data)} resumes.")
print(f"✅ Total Labels: {len(label2id)}")
```

```python
import json
import numpy as np
import evaluate
from datasets import load_from_disk
from transformers import (
    AutoModelForTokenClassification,
    TrainingArguments,
    Trainer,
    DataCollatorForTokenClassification,
    AutoTokenizer
)


with open("ner_config_bio.json", "r") as f:
    config = json.load(f)

label2id = config["label2id"]
id2label = {int(k): v for k, v in config["id2label"].items()}
MODEL_NAME = config["model"]

dataset = load_from_disk("processed_resume_dataset_bio")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

# 2. Define Metrics (using seqeval for NER)
metric = evaluate.load("seqeval")
label_list = [id2label[i] for i in range(len(id2label))]

def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=2)

    # Remove ignored index -100 and convert to label names
    true_predictions = [
        [label_list[p] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    true_labels = [
        [label_list[l] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]

    results = metric.compute(predictions=true_predictions, references=true_labels)
    return {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
        "accuracy": results["overall_accuracy"],
    }

# 3. Initialize Model
model = AutoModelForTokenClassification.from_pretrained(
    MODEL_NAME,
    num_labels=len(label2id),
    id2label=id2label,
    label2id=label2id
)

# 4. Training Arguments
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./bert_resume_ner_final",
    eval_strategy="epoch",           
    save_strategy="epoch",
    learning_rate=2e-5,               # Standard fine-tuning rate
    per_device_train_batch_size=16,   
    per_device_eval_batch_size=16,
    num_train_epochs=10,            
    weight_decay=0.01,                # Helps prevent overfitting
    warmup_steps=100,                 # Smooth start for the first few batches
    logging_dir='./logs',
    logging_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="f1",       # Always optimize for F1 in NER tasks
    fp16=True,                        
    report_to="none"                  
)

# 5. Data Collator (Handles dynamic padding)
data_collator = DataCollatorForTokenClassification(tokenizer)

# 6. Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["test"],
    processing_class=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

# 7. Start Training
print(" Starting training...")
trainer.train()

# 8. Save the Final Model
trainer.save_model("./final_ner_model")
print(" Model saved to ./final_ner_model")
```

```python
from transformers import pipeline, AutoTokenizer, AutoModelForTokenClassification
import json

# 1. Load the model and tokenizer
model_path = "./final_ner_model" 
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForTokenClassification.from_pretrained(model_path)


nlp_ner = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")


resume_text = """
SAKETH KUMAR
B.Tech in Computer Science from NIT Andhra Pradesh.
Skilled in Python, React, and Deep Learning.
Worked on an Examination Automation System using Django.
"""

# 4. Run Inference
results = nlp_ner(resume_text)

# 5. Print Results
print(f"{'Entity':<15} | {'Label':<10} | {'Score':<10}")
print("-" * 40)
for entity in results:
    print(f"{entity['word']:<15} | {entity['entity_group']:<10} | {entity['score']:.4f}")
```
